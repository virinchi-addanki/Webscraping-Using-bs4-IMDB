### Top 250 Rated Movies

```python
import openpyxl 
excel = openpyxl.Workbook()
sheet = excel.active
sheet.title = "Top 250 Rated Movies"
sheet.append(['ranking','name','year','rating','Number_of_ratings'])
```

```python
import requests,openpyxl
from bs4 import BeautifulSoup
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36 Edg/116.0.1938.81'}
url="https://www.imdb.com/chart/top/"
response = requests.get(url,headers=headers)
soup = BeautifulSoup(response.text,'html.parser')
print(soup)
```

```python
movies = soup.find('ul',class_='ipc-metadata-list ipc-metadata-list--dividers-between sc-3f13560f-0 sTTRj compact-list-view ipc-metadata-list--base').find_all('li')
#print(len(movies))
for movie in movies:
    ranking = movie.find('div',class_="sc-b51a3d33-0 hKhnaG cli-children").a.text.split('.')[0]
    name = movie.find('div',class_="sc-b51a3d33-0 hKhnaG cli-children").a.text.split('.')[1].lstrip()
    year = movie.find('div',class_="sc-b51a3d33-0 hKhnaG cli-children").span.text
    no_of_hours = movie.find('div',class_="sc-b51a3d33-5 ibuRZu cli-title-metadata").text.split(year[:len(year)])[1]
    rate= movie.find('span',class_="ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating").text
    rating =rate[:3]
    Number_of_ratings =rate[4:].strip('()')
    print(ranking,name,year,rating,Number_of_ratings)
    sheet.append([ranking,name,year,rating,Number_of_ratings])
excel.save('IMDB Movie Rating.xlsx')
    
    
```
